```python
from skimage.segmentation import slic, mark_boundaries
from skimage.data import astronaut
```

```python
import matplotlib.pyplot as plt
```

```python
img = astronaut()
```

```python
print img.shape
```

```python
plt.imshow(mark_boundaries(img, slic(img, 40, 10, sigma=1, enforce_connectivity=False)))
plt.title('skimage version\n enforce_connectivity=False', size=9)
plt.show()

```

```python
plt.imshow(mark_boundaries(img, slic(img, 40, 10, sigma=1, enforce_connectivity=True)))
plt.title('skimage version\n  enforce_connectivity=True', size=9)
plt.show()
```

```python
print img.dtype
```

```python
segments = slic(img, 100, 10, sigma=1, enforce_connectivity=True)
```

```python
plt.imshow(mark_boundaries(img, segments))
plt.show()
```
